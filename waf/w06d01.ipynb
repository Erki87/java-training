- Flexbox
	- Layout model
	- Responsive Web Design
	- Flow of content
	- Flex container
	- Flex items
	- One dimensional
		- Main direction
			- Items are arranged according to the main direction
			- Items do not stretch, but can shrink
		- Cross direction
			- Items stretch to fill the cross direction
