**Unit testing**
- Validate functionality
- Automated test of a "unit" (mostly individual method(s) - class)
- Reduce risk to introduce unwanted side effects
- Unit test -> Integration test -> System test -> Acceptance test
- Cheap and fast
- Helps with refactoring
- Usually only public methods
- Dependencies should be mocked/stubbed/etc

**Junit**

- Standard java framework for unit testing
- Absence of failure leads to success
- fail()
- Annotations
    - @Test
    - @Ignore
    - @Before
        - Runs before each test
    - @After
        - Runs after each test
    - @BeforeClass
        - Runs before unit test
    - @AfterClass
        - Runs after unit test
        
- Asserts
    - assertEquals()
    - assertTrue()
    - assertFalse()
    - assertNotNull()
    - assertNull()
    - assertArrayEquals()
