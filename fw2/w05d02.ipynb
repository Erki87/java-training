**Mockito**

- Java lib for creating mocks, stubs, etc
- Mockito does not differentiate between mocks and stubs
- https://www.vogella.com/tutorials/Mockito/article.html
- Methods:
    - mock()
    - when()
    - any()
    - isA()
    - thenReturn()    
    - thenThrow()
    - verify()
        - times()
        - never()
        - atLeastOnce()
        - atLeast()
        - atMost()
    - doReturn()
    - doNothing()
- Spies
    - https://www.baeldung.com/mockito-spy
    - http://sangsoonam.github.io/2019/02/04/mockito-doreturn-vs-thenreturn.html
- Argument captor
- Annotations

- Anitpatterns
    - https://rolf-engelhard.de/2013/04/logging-anti-patterns-part-ii/
