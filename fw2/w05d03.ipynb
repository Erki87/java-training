**Java EE**
- Java Platform, Enterprise Edition
- Specification for development of distributed computing and web applications
- Web server vs application server
- Applications
    - Web Application
        - Servlets (dynamic web content)
        - Static web content
        - Web Archive (WAR) file
            - classes
            - libs
            - web.xml deployment descriptor
    - EJB Application (Enterprise Java Beans)
        - Statless or stateful
        - Message driven (async, eg Java Message Service, MQSeries)
        - Business logic            
        - Full Java EE Application Server required
        - Java Archive (JAR) file
    - Enterprise Application
        - Wrapper for multiple Web and EJB applications
        - Full Java EE Application Server required
        - Enterprise Archive (EAR) file
            - universal classes
            - universal libs
            - multiple Java EE modules (eg Web Applications)           
            - application.xml deployment descriptor
- https://en.wikipedia.org/wiki/Java_Platform,_Enterprise_Edition

**Servlets**
- Specified in JEE
- Servlets generate dynamic web pages (like php)
    - Either via raw java code
    - Or using JSP (JavaServer Pages) or other technologies
- Hosted on application servers
- Sites are rendered on the server
    - Outdated to some extent
        - Most modern web sites use client side redering (js)
        - The data comes from rest services    
- Components
    - Servlet container/web container
        - Implemented by application server
        - Maps requests to servlets (web.xml from web app)
        - Creates and destroys servlets (web app)
    - Servlet
        - Managed by servlet container
        - Implemented by developer
        - Handles requests
        - Generates responses
        - Handles multiple threads
        - Interface javax.servlet.Servlet
            - https://docs.oracle.com/javaee/7/api/javax/servlet/Servlet.html
            - Not limited to http
            - init(), service() and destroy()
        - Abstract class javax.servlet.http.HttpServlet
            - Handles http requests and generates responses
            - https://docs.oracle.com/javaee/7/api/javax/servlet/http/HttpServlet.html            
- Life cycle
    - Client visits URL
        - Browser generates HTTP request
        - Request is sent app server
    - HTTP request is received by the server and forwarded to the servlet container
        - Servlet container maps request to particular servlet
        - Servlet container loads servlet if not yet loaded
            - Servlet container invokes init() method of the servlet.
    - Servlet container invokes the service() method of the servlet
        - Servlet reads data from request
        - Servlet generates response
    - Servlet container may unload servlet if unused
        - Servlet container calls destroy() method of the servlet
        - Servlet is garbage collected at some point
